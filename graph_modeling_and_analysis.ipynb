Vediamo come varia l'ascolto della canzoni in base al livello di energia di queste, durante le varie ore del giorno.

```python
MATCH (s:Song)-[:HAS_STREAMING_EVENT]->(e:StreamingEvent)
WHERE e.timestamp_utc IS NOT NULL
WITH apoc.date.parse(e.timestamp_utc, 'ms', 'MM/dd/yyyy hh:mm:ss a') AS timestamp_ms, s
WITH datetime({epochMillis: timestamp_ms}) AS timestamp_utc, s
WITH
  CASE
    WHEN time(timestamp_utc) >= time("06:00:00") AND time(timestamp_utc) < time("12:00:00") THEN '1 Mattina'
    WHEN time(timestamp_utc) >= time("12:00:00") AND time(timestamp_utc) < time("18:00:00") THEN '2 Pomeriggio'
    WHEN time(timestamp_utc) >= time("18:00:00") AND time(timestamp_utc) < time("00:00:00") THEN '3 Sera'
    WHEN time(timestamp_utc) >= time("00:00:00") AND time(timestamp_utc) < time("04:00:00") THEN '4 Notte'
ELSE '5 preMattina'
  END AS fascia_oraria,
  s.energy AS energy
WITH fascia_oraria, AVG(energy) AS media_energy
RETURN 'Energy per Fascia Oraria' AS index, fascia_oraria AS key, media_energy AS value
ORDER BY fascia_oraria;
```

```python
MATCH (s:Song)-[:HAS_STREAMING_EVENT]->(e:StreamingEvent)
WHERE e.timestamp_utc IS NOT NULL
WITH apoc.date.parse(e.timestamp_utc, 'ms', 'MM/dd/yyyy hh:mm:ss a') AS timestamp_ms, s
WITH datetime({epochMillis: timestamp_ms}) AS timestamp_utc, s
WITH
  CASE
    WHEN time(timestamp_utc) >= time("00:00:00") AND time(timestamp_utc) < time("01:00:00") THEN '00:00 - 01:00'
    WHEN time(timestamp_utc) >= time("01:00:00") AND time(timestamp_utc) < time("02:00:00") THEN '01:00 - 02:00'
    WHEN time(timestamp_utc) >= time("02:00:00") AND time(timestamp_utc) < time("03:00:00") THEN '02:00 - 03:00'
    WHEN time(timestamp_utc) >= time("03:00:00") AND time(timestamp_utc) < time("04:00:00") THEN '03:00 - 04:00'
    WHEN time(timestamp_utc) >= time("04:00:00") AND time(timestamp_utc) < time("05:00:00") THEN '04:00 - 05:00'
    WHEN time(timestamp_utc) >= time("05:00:00") AND time(timestamp_utc) < time("06:00:00") THEN '05:00 - 06:00'
    WHEN time(timestamp_utc) >= time("06:00:00") AND time(timestamp_utc) < time("07:00:00") THEN '06:00 - 07:00'
    WHEN time(timestamp_utc) >= time("07:00:00") AND time(timestamp_utc) < time("08:00:00") THEN '07:00 - 08:00'
    WHEN time(timestamp_utc) >= time("08:00:00") AND time(timestamp_utc) < time("09:00:00") THEN '08:00 - 09:00'
    WHEN time(timestamp_utc) >= time("09:00:00") AND time(timestamp_utc) < time("10:00:00") THEN '09:00 - 10:00'
    WHEN time(timestamp_utc) >= time("10:00:00") AND time(timestamp_utc) < time("11:00:00") THEN '10:00 - 11:00'
    WHEN time(timestamp_utc) >= time("11:00:00") AND time(timestamp_utc) < time("12:00:00") THEN '11:00 - 12:00'
    WHEN time(timestamp_utc) >= time("12:00:00") AND time(timestamp_utc) < time("13:00:00") THEN '12:00 - 13:00'
    WHEN time(timestamp_utc) >= time("13:00:00") AND time(timestamp_utc) < time("14:00:00") THEN '13:00 - 14:00'
    WHEN time(timestamp_utc) >= time("14:00:00") AND time(timestamp_utc) < time("15:00:00") THEN '14:00 - 15:00'
    WHEN time(timestamp_utc) >= time("15:00:00") AND time(timestamp_utc) < time("16:00:00") THEN '15:00 - 16:00'
    WHEN time(timestamp_utc) >= time("16:00:00") AND time(timestamp_utc) < time("17:00:00") THEN '16:00 - 17:00'
    WHEN time(timestamp_utc) >= time("17:00:00") AND time(timestamp_utc) < time("18:00:00") THEN '17:00 - 18:00'
    WHEN time(timestamp_utc) >= time("18:00:00") AND time(timestamp_utc) < time("19:00:00") THEN '18:00 - 19:00'
    WHEN time(timestamp_utc) >= time("19:00:00") AND time(timestamp_utc) < time("20:00:00") THEN '19:00 - 20:00'
    WHEN time(timestamp_utc) >= time("20:00:00") AND time(timestamp_utc) < time("21:00:00") THEN '20:00 - 21:00'
    WHEN time(timestamp_utc) >= time("21:00:00") AND time(timestamp_utc) < time("22:00:00") THEN '21:00 - 22:00'
    WHEN time(timestamp_utc) >= time("22:00:00") AND time(timestamp_utc) < time("23:00:00") THEN '22:00 - 23:00'
    ELSE '23:00 - 00:00'
  END AS fascia_oraria,
  s.energy AS energy
WITH fascia_oraria, AVG(energy) AS media_energy
RETURN 'Energy level avg per hour' AS index, fascia_oraria AS key, media_energy AS value
ORDER BY fascia_oraria;
```

```python
MATCH (s:Song)-[:HAS_STREAMING_EVENT]->(e:StreamingEvent)
WHERE s.title CONTAINS "Christmas" OR s.title CONTAINS "Xmas"
AND e.timestamp_utc IS NOT NULL
WITH apoc.date.parse(e.timestamp_utc, 'ms', 'MM/dd/yyyy hh:mm:ss a') AS timestamp_ms, s,e
WITH datetime({epochMillis: timestamp_ms}) AS timestamp_utc, s,e
WITH date(timestamp_utc) AS day, COUNT(e) AS value
RETURN day, value;
```

# Recomendation system

Questa query mira a generare delle raccomandazioni di canzoni non ancora ascoltate (che non hanno eventi di streaming associati), in base al genere delle canzoni che hanno già eventi di streaming.

```python
MATCH (s:Song)-[:HAS_STREAMING_EVENT]->(e:StreamingEvent),
      (s)-[:HAS_GENRE]->(g:Genre)<-[:HAS_GENRE]-(rec:Song)
WITH rec,g
Where NOT (rec)-[:HAS_STREAMING_EVENT]->()
WITH rec, g, COUNT(rec) AS countRec
RETURN rec.title AS RecommendedSong, g.name AS Genre, countRec
ORDER BY countRec DESC
LIMIT 100;
```
